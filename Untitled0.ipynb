###Predicting Employee Retention Rate Based On Some Factors

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python

from google.colab import files
df = pd.read_csv('HR_comma_sep.csv')
df.head(10)
```

**Satisfaction Level**: Satisfaction level seems to be relatively low (0.44) in employees leaving the firm vs the retained ones (0.66) </br>
**Average Monthly Hours**: Average monthly hours are higher in employees leaving the firm (199 vs 207) </br>
**Promotion Last 5 Years**: Employees who are given promotion are likely to be retained at firm

```python

salary = ['low', 'medium', 'high']
enc = OrdinalEncoder(categories=[salary])
enc.fit_transform(df[['salary']])
```

```python
ohenc = OneHotEncoder()
dpt = ohenc.fit_transform(df[['Department']]).toarray()
dpt_ctg = ohenc.categories_
dpt_ctg = np.array(dpt_ctg).ravel()
new_dpt = pd.DataFrame(dpt, columns=dpt_ctg)
merged = pd.concat([df,new_dpt], axis=1)
final = merged.drop(['Department','IT','salary'], axis=1)
final.head()
```

```python

import warnings
warnings.filterwarnings('ignore')
X = final.drop('left', axis=1)
Y = final['left']
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=35)
# Trying to use the scaler to improve my accuracy
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model = LogisticRegression()
model.fit(X_train,y_train)
print(y_test[10:20])
M = model.predict(X_test)
print(M[10:20])
print("The accuracy is ", end="")
model.score(X_test,y_test)*100
```

Tried using GridSearchCV to improve the accuracy

```python
from sklearn.model_selection import GridSearchCV
param_grid_lr = {
    'max_iter': [20, 50, 100, 200, 500, 1000],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'class_weight': ['balanced']
}
X1_train,X1_test,y1_train,y1_test = train_test_split(X,Y,test_size=0.3)
logModel_grid = GridSearchCV(estimator=LogisticRegression(random_state=195),param_grid=param_grid_lr, verbose=1, cv=10, n_jobs=-1)
logModel_grid.fit(X1_train, y1_train)
print(logModel_grid.best_estimator_)
```

```python
logModel_grid.predict(X1_test)
logModel_grid.score(X1_test,y1_test)
```

```python
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(y_test,M)
print(CM)
plt.figure(figsize=(15,10))
sns.heatmap(CM,annot=True,cmap='BuPu')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
